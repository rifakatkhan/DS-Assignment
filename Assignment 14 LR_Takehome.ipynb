```python
grade_male_ave = sum(Th_df[Th_df.gender == 'M'].numeric_class)/float(len(Th_df[Th_df.gender == 'M']))
grade_female_ave = sum(Th_df[Th_df.gender == 'F'].numeric_class)/float(len(Th_df[Th_df.gender == 'F']))
```

```python
# Now lets look at nationality
nation = Th_df.NationalITy.unique()
nation_grades_ave = [sum(Th_df[Th_df.NationalITy == i].numeric_class)/float(len(Th_df[Th_df.NationalITy == i])) for i in nation]
ax = sns.barplot(x=nation, y=nation_grades_ave)
jordan_ave = sum(Th_df[Th_df.NationalITy == 'Jordan'].numeric_class)/float(len(Th_df[Th_df.NationalITy == 'Jordan']))
print('Jordan average: '+str(jordan_ave))
plt.xticks(rotation=90)
```

###### As it can be seen in bar plot Jordan is seventh country with average 2.09 so 'Jordan' has positive impact on these two students actually

```python
# Lets look at relation with family members
relation = Th_df.Relation.unique()
relation_grade_ave = [sum(Th_df[Th_df.Relation == i].numeric_class)/float(len(Th_df[Th_df.Relation == i])) for i in relation]
ax = sns.barplot(x=relation, y=relation_grade_ave)
plt.title('Relation with father or mother affects success of students')
```

###### Having relation with mum has positive effect on these students

### 5. Build classification model and present it's classification report ?

```python
Th_df.head()
```

```python
Th_df1 = Th_df.drop('Class',axis = 1)
data_with_dummies = pd.get_dummies(Th_df1, drop_first=True)
```

```python
data_with_dummies.head()
```

```python
Features = data_with_dummies.drop(['numeric_class'],axis = 1)
Target = data_with_dummies['numeric_class']
```

```python
from sklearn.preprocessing import StandardScaler

scaler = StandardScaler()
scaler.fit(Features)
```

```python
X = scaler.fit_transform(Features)
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, Target, test_size=0.4, random_state=40)
```

```python
Logit_Model = LogisticRegression()
Logit_Model.fit(X_train,y_train)
```

```python
Prediction = Logit_Model.predict(X_test)
Score = accuracy_score(y_test,Prediction)
Report = classification_report(y_test,Prediction)
```
